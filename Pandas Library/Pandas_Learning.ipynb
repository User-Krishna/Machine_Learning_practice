# #Filtering dataframe on condition|masking using boolean series

```python
mask=data['city']=='Hyderabad'
data[mask]
```

```python
# define function one time and call it for any city to ensure how many match they had played in that city
def count_played_match(city):
    mask=data['city']==city
    return data[mask].shape[0]  #shape[0] is used to show rows only
```

```python
count_played_match('Hyderabad')
```

```python
count_played_match('Rajkot')
```

```python
#checking condition that list of matches that has played in hyderabad from 2010 onwards
def match_played_city_date(data, city_name):
    data['date'] = pd.to_datetime(data['date'])
    mask1 = data['city'] == city_name
    mask2 = data['date'] > '2010-01-01'
    return data[mask1 & mask2].shape[0]
```

```python
match_played_city_date(data,'Hyderabad')
```

# *group_by() functions on IPL dataset

```python
ipl=pd.read_csv('deliveries.csv')
ipl
```

```python
player=ipl.groupby('batsman')
```

```python
#find the top 10 batsman who had scored more runs in the ipl
player['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the top 5 batsman who had scored more number of fours
def batter_records(value):
    mask = ipl['batsman_runs'] == value
    return ipl[mask]
```

```python
top_fours = (
    batter_records(4)                      # filter only 4s
    .groupby('batsman')['batsman_runs']    # group by batsman
    .count()                               # count number of 4s
    .sort_values(ascending=False)          # sort descending
    .head(5)                               # top 5
)
print(top_fours)
```

```python
# Which batsman did score the most runs against opponents
def batsman_scored_against_others(batsman_name):
    return ipl[ipl['batsman']==batsman_name]  # masking step
```

```python
runs=(
    batsman_scored_against_others('MS Dhoni')
    .groupby('bowling_team')['batsman_runs']
    .sum()
    .sort_values(ascending=False)
    .head(1)
)
print(runs)
```

# #isin() function

```python
# Find the most destructive death over batsman in the history of IPL
# Strike rate=(Number of runs/Number of balls)*100
# Minimum batsman 200 balls in over 16-20

death_over=ipl[ipl['over']>15]
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
final = death_over[death_over['batsman'].isin(batsman_list)]
runs=final.groupby('batsman')['batsman_runs'].sum()
balls=final.groupby('batsman')['batsman_runs'].count()
sr=(runs/balls)*100
sr.sort_values(ascending=False).head(10)
```

# #Merge Functions

```python
#Finding which batsman won the Orange Cap and how many runs he scored.
new=ipl.merge(data,left_on='match_id',right_on='id') # merging two different dataset to get result. Here, left_on is performed on ipl dataframe and right_on on 'data' dataframe
```

```python
result = (
    new.groupby(['season','batsman'])['batsman_runs']
       .sum()
       .sort_values(ascending=False)
       .reset_index()
       .drop_duplicates(subset='season', keep='first')
       .sort_values('season')
       .reset_index(drop=True)
)

print(result)
```
